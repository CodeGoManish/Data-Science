```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Set plot title and labels
ax.set_title('Interactive Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Enable interactive mode
plt.ion()

# Show the plot
plt.show()

# Allow user to interact with the plot
while True:
    # Update data
    x.append(len(x) + 1)
    y.append(y[-1] + 2)

    # Clear previous plot
    ax.clear()

    # Plot updated data
    ax.plot(x, y)

    # Set plot title and labels
    ax.set_title('Interactive Plot')
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')

    # Pause to allow plot update to be visible
    plt.pause(0.5)
```
