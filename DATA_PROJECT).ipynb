```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Boston Housing Price dataset
boston_dataset = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")

# Perform data cleaning and filtering
# Check for missing values
print("Missing values:", boston_dataset.isnull().sum().sum())

# Filter rows with missing values
boston_dataset = boston_dataset.dropna()

# Perform descriptive and summary statistics
print("\nDescriptive Statistics:")
print(boston_dataset.describe())

# Plot distplot for target variable (medv)
sns.distplot(boston_dataset['medv'], kde=False)
plt.show()

# Plot heatmap for the correlation in the dataset
corr = boston_dataset.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python

import pandas as pd

# Load the Boston Housing Price dataset
boston_dataset = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")

# Group the data using index in pivot table
boston_grouped = boston_dataset.pivot_table(index='rad', values=['medv', 'crim', 'dis'])

# Aggregate on specific features with values
boston_agg = boston_grouped.agg(['mean', 'median'])

print(boston_agg)
```
